```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import sys
import requests
from selenium import webdriver
import time

sys.path.insert(0,'/content/chrome-win64')

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

driver = webdriver.Chrome(options=chrome_options)
print(driver)
```

```python
#create agent with mozilla header to avoid being blocked
headers = {
    "User-Agent": "Mozilla/5.0"
}
```

```python
#assign URL to variables
base_url = "https://stackoverflow.com"
user_url = f"{base_url}/users/22656/jon-skeet?tab=questions"

#view url
print(user_url)
```

```python
#send request
response = requests.get(user_url, headers=headers)

#conditional statements and raise exception
if response.status_code == 200:
    print("Success")
else:
  raise Exception(f"Failed to fetch page: {response.status_code}")

time.sleep(5)
```

```python
#pase the HTML
soup = BeautifulSoup(response.text, "html.parser")

#print
print(soup)
```

```python
#find all question summaries
questions = soup.select("div.js-post-summary")

data = []
for q in questions:
    title_tag = q.select_one("h3 a")
    vote_tag = q.select_one(".s-post-summary--stats-item__emphasized .s-post-summary--stats-item-number")

    if title_tag and vote_tag:
        title = title_tag.text.strip()
        link = base_url + title_tag.get('href')
        votes = vote_tag.text.strip()
        data.append([title, link, votes])

print(data)
```

```python
#create DataFrame
df = pd.DataFrame(data, columns=["Title", "Link", "Votes"])
print(df)
```

```python
#change votes to integers
#remove commas if any
df["Votes"] = df["Votes"].str.replace(",", "")
df["Votes"] = pd.to_numeric(df["Votes"], errors='coerce').fillna(0).astype(int)
```

```python
#mean max min and std metrics
mean_votes = df["Votes"].mean()
print(f"Mean Votes: {mean_votes:.2f}")

max_votes = df["Votes"].max()
print(f"Max Votes: {max_votes}")

min_votes = df["Votes"].min()
print(f"Min Votes: {min_votes}")

std_votes = df["Votes"].std()
print(f"Standard Deviation: {std_votes:.2f}")
```

```python
#normalizaiton of min-max
df["Votes_Normalized"] = (df["Votes"] - min_votes) / (max_votes - min_votes)
print(df)
```

```python
#filter questions with votes
above_mean = df[df["Votes"] > mean_votes]
long_titles = df[df["Title"].str.len() > 60]

#calc and add title a column for title length
long_titles["Title Length"] = long_titles["Title"].str.len()

print("\nQuestions with above-average votes:")
print(above_mean[["Title", "Votes"]])

print("\nQuestions with titles longer than 60 characters:")
print(long_titles[["Title", "Title Length"]])
```

```python
#show top 5 results
top_5 = df.sort_values(by="Votes", ascending=False).head(5)

print("Top 5 questions saved")
print(top_5)
```
